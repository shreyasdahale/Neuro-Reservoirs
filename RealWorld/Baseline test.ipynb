```python
import pickle

# Replace 'your_file.pkl' with your actual file name
with open('MIT-BIH Baseline results/mitbih.pkl', 'rb') as f:
    nrmse_dict = pickle.load(f)

# Now you can use the dictionary
print(nrmse_dict)
```

```python
nrmse_dict
```

```python
horizons = [300, 600, 1000]
```

```python
import numpy as np  # in case not already imported

model_names = ['ESN', 'SCR', 'CRJ', 'MCI-ESN', 'DeepESN']

print("\nNRMSE for Different Prediction Horizons:")
print("-" * 140)
header = "Horizon".ljust(10) + "".join([model.ljust(18) for model in model_names])
print(header)
print("-" * 140)

for horizon in horizons:
    row = f"{str(horizon):<10}"
    for model in model_names:
        model_vals = [np.mean(run[horizon]) for run in nrmse_dict[model]]
        mean = np.mean(model_vals)
        std = np.std(model_vals)
        row += f"{mean:.3f} ± {std:.3f}".ljust(18)
    print(row)
```
